## Conversión de datos
Vamos a transformar los datos que tenemos en números, para que pueda ser utilizado por nuestro algoritmo.

Empezaremos con el club, para ello vamos a transformar el String del nombre en un 1/0 usando una tabla.

```python
clb = df.pop("Club")

df = pd.concat([df.reset_index(drop=True), pd.get_dummies(clb, prefix='clb').reset_index(drop=True)], axis=1, sort=False)
df.head()
```

Hacemos lo mismo con Nationality y Position

```python
nac = df.pop("Nationality")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(nac, prefix='nac').reset_index(drop=True)], axis=1, sort=False)

pos = df.pop("Position")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(pos, prefix='pos').reset_index(drop=True)], axis=1, sort=False)

df.head()
```

Después de convertir las columnas con Strings, vamos a asegurarnos que el resto de valores sean numericos.
Como podemos ver arriba, los que no cumple dicha condicion son: **Value**, **Wage** y **LS**...**RB**

Primero de todo, vamos a crear una función para realizar esta conversion

```python
def value_to_float(x):
    """
    From K, M and + to float.
    
    """
    x = x.replace('€', '')
    ret_val = 0.0
    
    if type(x) == float or type(x) == int:
        ret_val = x
        
    if 'K' in x:
        if len(x) > 1:
            ret_val = float(x.replace('K', ''))
        ret_val = ret_val *1000
        
    if 'M' in x:
        if len(x) > 1:
            ret_val = float(x.replace('M', ''))
        ret_val = ret_val * 1000000.0  
    if '+1' in x:
        if len(x) > 1:
            ret_val = float(x.replace('+1', ''))
        ret_val = ret_val + 1.0
        
    if '+2' in x:
        if len(x) > 1:
            ret_val = float(x.replace('+2', ''))
        ret_val = ret_val + 2.0  
        
    if '+3' in x:
        if len(x) > 1:
            ret_val = float(x.replace('+3', ''))
        ret_val = ret_val + 3.0   
        
    return ret_val
```

Una vez creada, la aplicamos

```python
df["Value"] = df["Value"].apply(value_to_float)
df["Wage"] = df["Wage"].apply(value_to_float)
df["LS"] = df["LS"].apply(value_to_float)
df["ST"] = df["ST"].apply(value_to_float)
df["RS"] = df["RS"].apply(value_to_float)
df["LW"] = df["LW"].apply(value_to_float)
df["LF"] = df["LF"].apply(value_to_float)
df["CF"] = df["CF"].apply(value_to_float)
df["RF"] = df["RF"].apply(value_to_float)
df["RW"] = df["RW"].apply(value_to_float)
df["LAM"] = df["LAM"].apply(value_to_float)
df["CAM"] = df["CAM"].apply(value_to_float)
df["RAM"] = df["RAM"].apply(value_to_float)
df["LM"] = df["LM"].apply(value_to_float)
df["LCM"] = df["LCM"].apply(value_to_float)
df["CM"] = df["CM"].apply(value_to_float)
df["RCM"] = df["RCM"].apply(value_to_float)
df["RM"] = df["RM"].apply(value_to_float)
df["LWB"] = df["LWB"].apply(value_to_float)
df["LDM"] = df["LDM"].apply(value_to_float)
df["CDM"] = df["CDM"].apply(value_to_float)
df["RDM"] = df["RDM"].apply(value_to_float)
df["RWB"] = df["RWB"].apply(value_to_float)
df["LB"] = df["LB"].apply(value_to_float)
df["LCB"] = df["LCB"].apply(value_to_float)
df["CB"] = df["CB"].apply(value_to_float)
df["RCB"] = df["RCB"].apply(value_to_float)
df["RB"] = df["RB"].apply(value_to_float)

df.head()
```

## Predicion

Finalmente vamos a intentar predecir el **value**

```python
val = df.pop("Value")

df
```

Tenemos los datos que queremos predecir en formato numérico, la información del valor. Entonces ahora podemos usar estos datos para entrenar un modelo. Pero primero necesitamos dividir los datos para poder conocer su verdadero desempeño.

```python
X_train, X_test, y_train, y_test = train_test_split(df, val, test_size=0.33, random_state=50)
```

```python
len(X_train)
```

Entrenamos nuestro modelo de LinearRegression.

```python
reg = linear_model.LinearRegression().fit(X_train, y_train)
```

Finalmente obtenemos una métrica $ R ^ 2 $ para la regresión, usamos la implementación de <a href="https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html"> sickit-learn</a>.

```python
preds = reg.predict(X_test)
```

```python
r2_score(preds, y_test)
```

## Resultado
Como podemos ver tenemos un $ R ^ 2 $ score de +0.8, siendo el ideal de 1.0
